Unter Normaldruck bei $20°C$ beträgt der theoretische Adiabatenexponent laut [Wikipedia](https://www.wikipedia.org/wiki/Isotropenexponent)
$$\gamma_{thor} =  1.40$$

Im Raum wo das Experiment stattfand, exestierten zu dem Zeitpunkt des Experiments sehr ähnliche Druck und Termperatur verhältnisse, so dass Normalbedingungen angeommen werden können. Weshalb sehr ähnliche Werte bei unseren Versuchen rausgekommen sind.
