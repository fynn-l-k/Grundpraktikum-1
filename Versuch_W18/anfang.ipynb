

# Verscuhsaufbau (Aufgabe 2)
In Aufgabe 2 musste eine Ferromagnet in einem dünnen Rohr mit Hilfe einer Luftpumpe zum schweben gebracht. Der Aufbau bestand aus einem dünnen Glasrohr mit zwei öffnungne oben und unten um den Druck mittels einer Fahrradluftpumpe einzustellen. In der Glasrohre war ein genauso dicker Magnetzylinder und in der Mitte der Röhre wurde sie durch eine Spule umgeben. Wenn der Ferromagnet bei der Spule angekommen ist, musst die Frequenz so eingestellt werden, das es uzum Resonanzfall kam. So wurde optisch geschaut wann der Magnet am meisten schwingt. Dfür wurde der Versuch mehrmals wiederholt, da die Wechselstromquelle, sehr ungenau war.
