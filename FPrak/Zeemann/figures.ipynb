```python
from matplotlib import rcParams
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.style as mtlstyle

from scipy.optimize import curve_fit

mtlstyle.use("seaborn-v0_8-paper")

rcParams["figure.dpi"] = 300
rcParams["figure.figsize"] = (5, 3)

```

```python


# CSV einlesen
df = pd.read_csv('Aufgabe_1_1.csv')
df.columns = df.columns.str.replace('"', '').str.strip()

def f(I, a, b):
    return a * I + b

popt, pcov = curve_fit(f, df['I[A]'], df['B[mT]'])

print(popt)

# Plot für a_x
plt.figure(figsize=(5, 3))
plt.scatter(df['I[A]'], df['B[mT]'], label="Magnetic field strength", color="blue", marker='o')
plt.plot(df['I[A]'], f(df['I[A]'], *popt), label=f"Calibration function\n$B={popt[0]:.2f}±{np.sqrt(pcov[0,0]):.2f} ~ I + {popt[1]:.2f}±{np.sqrt(pcov[1,1]):.2f}$", color="orange")
plt.xlabel("Current [$A$]")
plt.ylabel("Magnetic field strength [$mT$]")
plt.title("Magnetic field strength against the current")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ============================================================
# CONSTANTS
# ============================================================

h = 6.62607015e-34          # Planck constant [J s]
c = 2.99792458e8            # Speed of light [m/s]
d = 0.003                   # Fabry-Perot plate distance [m]

# Measurement uncertainties
delta_D = 1.0               # Diameter uncertainty [px]
rel_B_error = 0.01          # Relative magnetic field uncertainty

# ============================================================
# MEASUREMENT DATA
# ============================================================

B_mT = np.array([
    173.61,
    251.67,
    329.73,
    407.79,
    485.85,
    563.91,
    641.97,
    720.03,
    798.09
])

D11 = np.array([122,121,117,114,111,107,105,102,98], dtype=float)
D31 = np.array([136,138,140,143,146,150,152,154,156], dtype=float)

D12 = np.array([224,222,220,218,216,214,213,211,210], dtype=float)
D32 = np.array([229,231,233,235,238,239,241,241,243], dtype=float)

# Convert mT -> T
B = B_mT * 1e-3

# Magnetic field uncertainty
B_error = rel_B_error * B

# ============================================================
# ENERGY SPLITTING
# ============================================================

prefactor = (h * c) / (4 * d)

def delta_E(D11, D12, D31, D32):

    inner = (D11**2) / (D12**2 - D11**2)
    outer = (D31**2) / (D32**2 - D31**2)

    return prefactor * np.abs(inner - outer)

# ============================================================
# NUMERICAL ERROR PROPAGATION
# ============================================================

def numerical_error(D11, D12, D31, D32, delta=1.0):

    variables = [D11, D12, D31, D32]
    derivatives = []

    for i in range(4):

        plus = variables.copy()
        minus = variables.copy()

        plus[i] += delta
        minus[i] -= delta

        f_plus = delta_E(*plus)
        f_minus = delta_E(*minus)

        derivative = (f_plus - f_minus) / (2 * delta)

        derivatives.append(derivative)

    derivatives = np.array(derivatives)

    sigma = np.sqrt(np.sum((derivatives * delta)**2))

    return sigma

# ============================================================
# CALCULATE ΔE
# ============================================================

DeltaE = []
DeltaE_error = []

for i in range(len(B)):

    de = delta_E(D11[i], D12[i], D31[i], D32[i])
    err = numerical_error(D11[i], D12[i], D31[i], D32[i])

    DeltaE.append(de)
    DeltaE_error.append(err)

DeltaE = np.array(DeltaE)
DeltaE_error = np.array(DeltaE_error)

# ============================================================
# LINEAR FIT
# ============================================================

def linear(x, m):
    return m * x

popt, pcov = curve_fit(
    linear,
    B,
    DeltaE,
    sigma=DeltaE_error,
    absolute_sigma=True
)

mu_B = popt[0]
mu_B_error = np.sqrt(np.diag(pcov))[0]

# ============================================================
# RESULTS TABLE
# ============================================================

df = pd.DataFrame({
    "B [mT]": B_mT,
    "D11 [px]": D11,
    "D31 [px]": D31,
    "D12 [px]": D12,
    "D32 [px]": D32,
    "Delta E [J]": DeltaE,
    "Error Delta E [J]": DeltaE_error
})

print("\n====================================================")
print("RESULTS")
print("====================================================\n")

print(df)

# ============================================================
# BOHR MAGNETON
# ============================================================

mu_B_lit = 9.2740100783e-24

relative_deviation = abs(mu_B - mu_B_lit) / mu_B_lit

print("\n====================================================")
print("BOHR MAGNETON")
print("====================================================\n")

print(f"Experimental value:")
print(f"μ_B = ({mu_B:.4e} ± {mu_B_error:.4e}) J/T")

print(f"\nLiterature value:")
print(f"μ_B,literature = {mu_B_lit:.4e} J/T")

print(f"\nRelative deviation:")
print(f"{relative_deviation:.2%}")

# ============================================================
# PLOT
# ============================================================

plt.figure(figsize=(9,6))

# Errorbar plot
plt.errorbar(
    B,
    DeltaE,
    xerr=B_error,
    yerr=DeltaE_error,
    fmt='o',
    markersize=6,
    capsize=4,
    elinewidth=1.2,
    label='Measured data'
)

# Fit line
xfit = np.linspace(0, max(B)*1.05, 500)

plt.plot(
    xfit,
    linear(xfit, mu_B),
    '--',
    linewidth=2,
    label=(
        r'Fit: $\mu_B = '
        f'({mu_B:.2e} \\pm {mu_B_error:.1e})'
        r'\ \mathrm{J/T}$'
    )
)

plt.xlabel(r'Magnetic flux density $B$ [T]', fontsize=13)
plt.ylabel(r'Energy splitting $\Delta E$ [J]', fontsize=13)

#plt.title('Normal Zeeman Effect', fontsize=15)

plt.grid(True, linestyle='--', alpha=0.6)

plt.legend(fontsize=11)

plt.tight_layout()
plt.show()

# ============================================================
# LATEX TABLE
# ============================================================

latex_df = df.copy()

latex_df["Delta E [J]"] = latex_df["Delta E [J]"].map(lambda x: f"{x:.4e}")
latex_df["Error Delta E [J]"] = latex_df["Error Delta E [J]"].map(lambda x: f"{x:.4e}")

print("\n====================================================")
print("LATEX TABLE")
print("====================================================\n")

print(latex_df.to_latex(index=False))
```
