# Plot Licht

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.style as mtlstyle

from scipy.optimize import curve_fit

mtlstyle.use("seaborn-v0_8-paper")

```

```python


# CSV einlesen
df = pd.read_csv('Aufgabe_1_1.csv')
df.columns = df.columns.str.replace('"', '').str.strip()

def f(I, a, b):
    return a * I + b

popt, pcov = curve_fit(f, df['I[A]'], df['B[mT]'])

print(popt)

# Plot für a_x
plt.figure(figsize=(5, 3))
plt.scatter(df['I[A]'], df['B[mT]'], label="Magnetic field strength", color="blue", marker='o')
plt.plot(df['I[A]'], f(df['I[A]'], *popt), label=f"Calibration function\n$B={popt[0]:.2f}±{np.sqrt(pcov[0,0]):.2f} ~ I + {popt[1]:.2f}±{np.sqrt(pcov[1,1]):.2f}$", color="orange")
plt.xlabel("Current [$A$]")
plt.ylabel("Magnetic field strength [$mT$]")
plt.title("Magnetic field strength against the current")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('Aufgabe_1_2_1A.csv')
df.columns = df.columns.str.replace('"', '').str.strip()

df_2 = pd.read_csv('Aufgabe_1_2_2A.csv')
df_2.columns = df_2.columns.str.replace('"', '').str.strip()


df_6 = pd.read_csv('Aufgabe_1_2_6A.csv')
df_6.columns = df_6.columns.str.replace('"', '').str.strip()

df_10 = pd.read_csv('Aufgabe_1_2_10A.csv')
df_10.columns = df_10.columns.str.replace('"', '').str.strip()

df_3 = pd.read_csv('Aufgabe_1_2_3A.csv')
df_3.columns = df_3.columns.str.replace('"', '').str.strip()

df_4 = pd.read_csv('Aufgabe_1_2_4A.csv')
df_4.columns = df_4.columns.str.replace('"', '').str.strip()

df_5 = pd.read_csv('Aufgabe_1_2_5A.csv')
df_5.columns = df_5.columns.str.replace('"', '').str.strip()

df_7 = pd.read_csv('Aufgabe_1_2_7A.csv')
df_7.columns = df_7.columns.str.replace('"', '').str.strip()

df_8 = pd.read_csv('Aufgabe_1_2_8A.csv')
df_8.columns = df_8.columns.str.replace('"', '').str.strip()

df_9 = pd.read_csv('Aufgabe_1_2_9A.csv')
df_9.columns = df_9.columns.str.replace('"', '').str.strip()

# Plot für a_x
plt.figure(figsize=(5, 3))
plt.scatter(df['s[mm]'], df['B[mT]'], label="Magnetic flux density", color="blue", marker='o')
plt.xlabel("Position [$mm$]")
plt.ylabel("Magnetic flux density [$mT$]")
plt.title("1 A Magnetic flux density über die Position")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()




plt.figure(figsize=(5, 3))
plt.scatter(df_2['s[mm]'], df_2['B[mT]'], label="Magnetic flux density", color="blue", marker='o')
plt.xlabel("Position [$mm$]")
plt.ylabel("Magnetic flux density [$mT$]")
plt.title("2 A Magnetic flux density über die Position")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()




plt.figure(figsize=(5, 3))
plt.scatter(df_6['s[mm]'], df_6['B[mT]'], label="Magnetic flux density", color="blue", marker='.')
plt.xlabel("Position [$mm$]")
plt.ylabel("Magnetic flux density [$mT$]")
plt.title("6 A Magnetic flux density über die Position")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


def f(I, a, b, c, d, e, f, g):
    return g * I**6 + a * I**5 + b * I **4 + c * I ** 3 + d * I ** 2 + e * I ** 1 + f

popt, pcov = curve_fit(f, df_10['s[mm]'], df_10['B[mT]'])

print(popt)

plt.figure(figsize=(5, 3))
plt.scatter(df_10['s[mm]'], df_10['B[mT]'], label="Magnet flux density", color="blue", marker='o')
plt.plot(df_10['s[mm]'], f(df_10['s[mm]'], *popt), label=f"Ploynomial fit", color="orange", marker=',')
plt.xlabel("Position [$mm$]")
plt.ylabel("Magnetic flux density [$mT$]")
plt.title("10 A Magnetic flux density against the postion of the Hall sensor")
plt.grid(True)
plt.legend(loc='lower center')
plt.tight_layout()
plt.show()


plt.figure(figsize=(5, 3))
plt.scatter(df['s[mm]'], df['B[mT]'], label="Magnetic flux density 1 A", color="blue", marker='o')
plt.scatter(df_2['s[mm]'], df_2['B[mT]'], label="Magnetic flux density 2 A", color="green", marker='o')
plt.scatter(df_3['s[mm]'], df_3['B[mT]'], label="Magnetic flux density 3 A", color="blue", marker='o')
plt.scatter(df_4['s[mm]'], df_4['B[mT]'], label="Magnetic flux density 4 A", color="green", marker='o')
plt.scatter(df_5['s[mm]'], df_5['B[mT]'], label="Magnetic flux density 5 A", color="orange", marker='o')
plt.scatter(df_6['s[mm]'], df_6['B[mT]'], label="Magnetic flux density 6 A", color="red", marker='o')
plt.scatter(df_7['s[mm]'], df_7['B[mT]'], label="Magnetic flux density 7 A", color="purple", marker='o')
plt.scatter(df_8['s[mm]'], df_8['B[mT]'], label="Magnetic flux density 8 A", color="brown", marker='o')
plt.scatter(df_9['s[mm]'], df_9['B[mT]'], label="Magnetic flux density 9 A", color="pink", marker='o')
plt.scatter(df_10['s[mm]'], df_10['B[mT]'], label="Magnetic flux density 10 A", color="black", marker='o')
plt.xlabel("Position [$mm$]")
plt.ylabel("Magnetic flux density [$mT$]")
plt.title("10 A Magnetic flux density über die Position")
plt.grid(True)
plt.legend(bbox_to_anchor=(0., -0.2, 1., -0.1), loc='upper left',
                      ncols=2, mode="expand", borderaxespad=0.)
plt.tight_layout()
plt.show()
```
