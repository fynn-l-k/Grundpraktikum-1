# Plot Licht

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('Aufgabe_1_1.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



# Plot für a_x
plt.figure(figsize=(10, 4))
plt.scatter(df['I[A]'], df['B[mT]'], label="Magnetfeldstärke", color="blue", marker='o')
plt.xlabel("Stromstärke (A)")
plt.ylabel("Magnetfeldstärke (mT)")
plt.title("Magnetfeldstärke über die Stromstärke")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
