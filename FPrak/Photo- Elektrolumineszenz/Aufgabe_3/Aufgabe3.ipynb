```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe3_blaueLED.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="blau LED", color="blue")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.axvline(473, color='gray', linestyle='--', label='473.0 nm')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('plot_aufgabe_3.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
