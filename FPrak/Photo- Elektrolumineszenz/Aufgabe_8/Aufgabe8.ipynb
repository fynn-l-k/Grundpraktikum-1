```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe8_AlGaAs_minus184_4K_1127.csv')
df2 = pd.read_csv('csv/Aufgabe8_AlGaAs_zimmertemp_1127.csv')
df.columns = df.columns.str.replace('"', '').str.strip()
df2.columns = df2.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="88.75 K", color="blue")
plt.plot(df2['Wellenlänge [nm]'], df2['Intensität [mV]'], label="293.15 K", color="lightcoral")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.grid(True)
plt.axvline(672.1, color='lightgray', linestyle='--', label='672.1 nm')
plt.axvline(690.5, color='darkgray', linestyle='--', label='690.5 nm')
plt.axvline(705, color='gray', linestyle='--', label='705.0 nm')
plt.legend()
plt.xlim(640, 750)
plt.ylim(-0.02, 0.33)
plt.tight_layout()

plt.savefig('plot_aufgabe_8_1127.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
