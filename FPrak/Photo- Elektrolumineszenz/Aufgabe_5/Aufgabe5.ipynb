```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe5_orangeneLED.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="orangene LED", color="orange")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.axvline(585.6, color='gray', linestyle='--', label='585.6 nm')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('plot_aufgabe_5.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
