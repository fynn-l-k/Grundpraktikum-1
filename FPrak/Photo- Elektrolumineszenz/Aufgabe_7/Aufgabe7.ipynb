```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe7_infraroteLED.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="infrarote LED", color="red")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.grid(True)
plt.axvline(899, color='gray', linestyle='--', label='899.0 nm')
plt.legend()
plt.tight_layout()

plt.savefig('plot_aufgabe_7.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
