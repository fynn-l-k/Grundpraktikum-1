```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df_2 = pd.read_csv('Aufgabe_2/csv/Aufgabe2_weißeLED.csv')
df_2.columns = df_2.columns.str.replace('"', '').str.strip()
df_3 = pd.read_csv('Aufgabe_3/csv/Aufgabe3_blaueLED.csv')
df_3.columns = df_3.columns.str.replace('"', '').str.strip()
df_4 = pd.read_csv('Aufgabe_4/csv/Aufgabe4_grueneLED.csv')
df_4.columns = df_4.columns.str.replace('"', '').str.strip()
df_5 = pd.read_csv('Aufgabe_5/csv/Aufgabe5_orangeneLED.csv')
df_5.columns = df_5.columns.str.replace('"', '').str.strip()
df_6 = pd.read_csv('Aufgabe_6/csv/Aufgabe6_roteLED.csv')
df_6.columns = df_6.columns.str.replace('"', '').str.strip()

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df_2['Wellenlänge [nm]'], df_2['Intensität [mV]'], label="weiße", color="black")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df_3['Wellenlänge [nm]'], df_3['Intensität [mV]'], label="Blaue LED", color="blue")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df_4['Wellenlänge [nm]'], df_4['Intensität [mV]'], label="Grüne LED", color="green")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df_5['Wellenlänge [nm]'], df_5['Intensität [mV]'], label="Orangene LED", color="orange")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df_6['Wellenlänge [nm]'], df_6['Intensität [mV]'], label="Rote LED", color="red")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot LEDS
plt.figure(figsize=(10, 4))
plt.plot(df_2['Wellenlänge [nm]'], df_2['Intensität [mV]'], label="Weiße LED", color="black")
plt.plot(df_3['Wellenlänge [nm]'], df_3['Intensität [mV]'], label="Blaue LED", color="blue")
plt.plot(df_4['Wellenlänge [nm]'], df_4['Intensität [mV]'], label="Grüne LED", color="green")
plt.plot(df_5['Wellenlänge [nm]'], df_5['Intensität [mV]'], label="Orangene LED", color="orange")
plt.plot(df_6['Wellenlänge [nm]'], df_6['Intensität [mV]'], label="Rote LED", color="red")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke der LEDs über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot LEDS
plt.figure(figsize=(10, 4))
plt.plot(df_3['Wellenlänge [nm]'], df_3['Intensität [mV]'], label="Blaue LED", color="blue")
plt.plot(df_4['Wellenlänge [nm]'], df_4['Intensität [mV]'], label="Grüne LED", color="green")
plt.plot(df_5['Wellenlänge [nm]'], df_5['Intensität [mV]'], label="Orangene LED", color="orange")
plt.plot(df_6['Wellenlänge [nm]'], df_6['Intensität [mV]'], label="Rote LED", color="red")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.title("Lichtstärke der LEDs über die Wellenlänge")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
