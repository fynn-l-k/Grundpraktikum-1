```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe4_grueneLED.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="grüne LED", color="green")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.axvline(557.5, color='gray', linestyle='--', label='557.5 nm')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('plot_aufgabe_4.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
