```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('csv/Aufgabe2_weißeLED.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



plt.figure(figsize=(7, 4))
plt.plot(df['Wellenlänge [nm]'], df['Intensität [mV]'], label="weiße LED", color="black")
plt.xlabel("Wellenlänge [nm]")
plt.ylabel("Intensität [mV]")
plt.axvline(469.5, color='gray', linestyle='--', label='469.5 nm')
plt.axvline(545, color='gray', linestyle='--', label='545.0 nm')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('plot_aufgabe_2.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
