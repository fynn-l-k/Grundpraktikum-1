Für Aufgabe 2 haben wir uns für den freine Fall als Bewegungsvorgang entscheiden.
Dabei analysieren wir folgende Punkte:
- Ermittlung der Geschwindigkeit
- Kann Luftwiderstand vernachlässigt werden?
- Wie viel kinetische Energie verliert der Körper beim Aufprall?

## Messung

Für die Messung wurde eine Ball von einer Höhe vomn 0.8m fallen gelassen. Dabei hatte der Ball eine Masse von 18g und einen Radius von 3.5cm. Während des falles wurde mit einer Kamera mit 240fps die Bewegung aufgenommen.
Als Bewegungsgleichung wird die Formel für den freien Fall verwendet:
$$ s(t) =\frac{1}{2}gt^2 $$

## Ermittlung der Geschwindigkeit
Die Geschwindigkeit wird mit der Formel $v(t) = \frac{ds}{dt}$ ermittelt. Dabei wird die Ableitung der Bewegungsgleichung $s(t)$ gebildet.

## Kann Luftwiderstand vernachlässigt werden?
Die Luftwiderstandskraft ist gegeben durch:
$$
F_{Luft} = \frac{1}{2} \cdot \rho \cdot v^2 \cdot A \cdot c_w
$$
so ergibt sich eine neue Bewegungsgleichung für den freien Fall:
$$
s(t) =\left( \frac{1}{2} g - \frac{1}{2} \frac{\rho}{m} A c_w \right) \cdot t^2
$$
