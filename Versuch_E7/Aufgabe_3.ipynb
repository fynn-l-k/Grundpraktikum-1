Spulenpaar 3 mit Abstand a = 0.5 R

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Konstanten
mu_0 = 4 * np.pi * 1e-7  # T·m/A
I = 0.93  # A
N = 320
R = 0.068  # m

# Funktion für theoretische Berechnung
def B_theor(z, a):
    term1 = mu_0 * I * N * R**2 / (2 * (R**2 + (z + a/2)**2)**1.5)
    term2 = mu_0 * I * N * R**2 / (2 * (R**2 + (z - a/2)**2)**1.5)
    return (term1 + term2) * 10**6

# CSV-Datei laden (bitte Pfad anpassen!)
df = pd.read_csv('data/Spulenpaar0_5R.csv')
df.columns = ['B_gemessen', 'z']

# Umrechnen von mm in m, falls nötig
if df['z'].max() > 1:
    df['z'] /= 1000

# Berechne theoretische Werte
df['B_theor'] = B_theor(df['z']-0.835, 0.5 * R)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(df['z'], df['B_gemessen'], 'o-', label='Gemessen')
plt.plot(df['z'], df['B_theor'], 'x--', label='Theorie')
plt.title('a = 0.5R')
plt.xlabel('z [m]')
plt.ylabel('B [mT]')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
