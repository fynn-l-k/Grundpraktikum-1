# Plot Licht

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('arduino_data_2.csv')
df.columns = df.columns.str.replace('"', '').str.strip()

# Zeitspalte erzeugen
df['Zeitstempel'] = pd.to_datetime(
    df['Jahr'].astype(str) + "-" +
    df['Monat'].astype(str).str.zfill(2) + "-" +
    df['Tag'].astype(str).str.zfill(2) + " " +
    df['Stunde'].astype(str).str.zfill(2) + ":" +
    df['Minute'].astype(str).str.zfill(2) + ":" +
    df['Sekunde'].astype(str).str.zfill(2),
    format='%Y-%m-%d %H:%M:%S'
)

# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df['Zeitstempel'], df['Lux (lx)'], label="Licht", color="blue")
plt.xlabel("Zeit (Monat-Tag Stunde)")
plt.ylabel("Lux (lx))")
plt.title("Lichtstärke über Zeit")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
