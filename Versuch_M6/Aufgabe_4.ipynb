```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('Daten/Aufgabe_4.csv')
df.columns = ['t[s]', 'RPM_AVG']


# Plot für die Messdaten
plt.figure(figsize=(10, 6))
plt.plot(df['t[s]'], df['RPM_AVG'],'o', label="Messdaten", color="blue",)
plt.xlabel("V [ml]")
plt.ylabel("p [$10^5$Pa]")
plt.title("Isotherme für 31,5 °C")
#plt.xscale('log')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
