```python
#Plotten der Daten
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('Aufgabe_1_Daten.csv')
df.columns = ['R_L[Ohm]', 'I[mA]', 'U[V]', 'Messbereich']



# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df['I[mA]'], df['U[V]'],'o', label="Lichtstärke", color="blue",)
plt.xlabel("Zeit (Monat-Tag Stunde)")
plt.ylabel("Lux (lx)")
plt.title("Lichtstärke über die Zeit")
#plt.xscale('log')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
