## Graphische Darstellung mit Fit
Wie in der [Abbildung @fig-aufgabe-1] zu sehen ist wurde eine curve fit über die Messdaten gelegt um den Innenwiederstand und die Quellenspannung zu erhalten. Da der Curve Fit, die lineare Beziehung zwischen Strom und Spannung beschreibt. Dabei kammen wir auf die folgende Werte:

|Gefittete Werte mit Unsicherheiten|
|-----------|-------------------------------|
| Innenwiederstand | $110.2012 ± 0.0471 \Omega$|
| Quellspannung | $3.0849 ± 0.0007 V$ |
: Tabelle mit den ermittelten Werte für den Innenwiderstand und die Quellenspannung aus dem Curve Fit (Minus weg, da negativer Wiederstand nicht möglich)


```python
# | fig-cap: Spannung über Strom mit einem Curve Fit
# | label: fig-aufgabe-1
#Plotten der Daten
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

# CSV einlesen
df = pd.read_csv('Daten/Aufgabe_1_Daten.csv')
df.columns = ['R_L[Ohm]', 'I[mA]', 'U[V]', 'Messbereich']

df['I[mA]'] = df['I[mA]'] * 1e-3  # Umwandlung von mA nach A

def func(x, a, b):
    return a*x +b

params, cov = curve_fit(func, df['I[mA]'], df['U[V]'])

I = np.linspace(0, np.max(df['I[mA]']), 1000)

#print("Gefittete Werte mit Unsicherheiten:", )
#print("Innenwiederstand: {:.4f} ± {:.4f} Ω".format(params[0], np.sqrt(cov[0][0])))
#print("Quellspannung: {:.4f} ± {:.4f} V".format(params[1], np.sqrt(cov[1][1])))

# Plot für a_x
plt.figure(figsize=(10, 6))
plt.plot(df['I[mA]'], df['U[V]'],'o', label="Messdaten", color="blue",)
plt.plot(I, func(I, *params), label="Linearer Fit", color="orange")
plt.xlabel("Strom [$A$]")
plt.ylabel("Spannung [$V$]")
plt.title("Spannung über den Strom")
#plt.xscale('log')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Bestimmung des Kurzschlussstroms $I_K$
Der Kurzschlussstrom lässt sich bestimmen mit der Formel
$$I_K = \frac{U_q}{R_I}$$
Aus dem [vorherigen Abschnitt](#graphische-darstellung-mit-fit) ergab sich für die Quellenspannung $U_q$ und den Innenwiderstand $R_I$ folgende Werte:
$R_I = 110.2012 \pm 0.0471  \Omega \text{ und } U_q = 3.0849 \pm 0.0007 V$$

Damit ergibt sich für den Kurzschlussstrom $I_K$:
$$I_K = \frac{3.0849 \pm 0.0007 V}{110.2012 \pm 0.0471 \Omega} \approx 0.0280 \pm 0.000014 A$$
