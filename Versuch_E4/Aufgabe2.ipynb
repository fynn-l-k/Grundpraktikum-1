## Graphische Darstellung als $P-R_a-\text{Plot}$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

df = pd.read_csv('./Daten/Aufgabe_1_Daten.csv')

df['I[mA]'] = df['I[mA]'] * 10 ** (-3)  # Convert current from mA to A
df['P[W]'] = df['I[mA]'] * df['U[V]']

#df.to_csv('./Daten/Aufgabe_1_Daten_mit_P.csv', index=False)

#print as csv
plt.figure(figsize=(10, 6))
plt.plot(df['R_L[Ohm]'], df['P[W]'], 'o')
plt.xlabel('R_L[Ohm]')
plt.ylabel('P[W]')
plt.vlines(x=110, ymin=0, ymax=np.max(df['P[W]']), color='orange', linestyle='--', label="Leistungspunkt, der Messwerte (110 Ohm)")
#print(np.max(df['P[W]']))
plt.title("Leistung $P$ über den Außenwiederstand $R_a$")
plt.grid()
plt.xscale('log')
plt.legend()
#plt.yscale('log')
plt.show()
```
