## Graphische Darstellung als $P-R_a-\text{Plot}$

```python
# | label: fig-aufgabe-2-1
# | fig-cap: Leistung über Lastwiderstand, mit einem quadratischen Fit im Peak Bereich für genauere Bestimmung des Maximums
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Quadratische Funktion
def parabola(R, a, b, c):
    return a * R**2 + b * R + c

# Bereich um den Peak wählen (z. B. 60 bis 150 Ohm)
df_peak = df[(df['R_L[Ohm]'] > 60) & (df['R_L[Ohm]'] < 150)]
R_peak = df_peak['R_L[Ohm]'].values
P_peak = df_peak['P[W]'].values

# Fit durchführen
params, _ = curve_fit(parabola, R_peak, P_peak)
a, b, c = params

# Maximum berechnen (Scheitelpunkt)
R_max = -b / (2 * a)
P_max = parabola(R_max, *params)

# Plot
R_fit = np.linspace(60, 180, 600)
P_fit = parabola(R_fit, *params)

plt.figure(figsize=(10, 6))
plt.plot(df['R_L[Ohm]'], df['P[W]'], 'o', label='Messdaten')
plt.plot(R_fit, P_fit, '--', label='Quadratischer Fit um Peak', color="red")
plt.axvline(R_max, color='gray', linestyle='--', label=f'Fit-Maximum: {R_max:.1f} Ω ({P_max:.7f} W)')
#plt.axhline(P_max, color='gray', linestyle=':', label=f'Max. Leistung: {P_max:.4f} W')
plt.vlines(x=110, ymin=0, ymax=np.max(df['P[W]']), color='orange', linestyle='--', label="Maximum der Messwerte: 110 Ω (0.0215968 W)")

plt.xlabel('$R_L$ [$\\Omega$]')
plt.ylabel('$P$ [W]')
plt.xscale('log')
plt.title("Quadratischer Fit im Peak-Bereich")
plt.legend()
plt.grid(True)
plt.show()

```

## Fit der Messdaten
Zur Auswertung der Messdaten wurde ein Curve Fit mit der folgenden Formel verwendet:
$$P(R_a) = \frac{U_q^2 R_a}{(R_a + R_I)^2}$$

Diese Formel beschreibt die Leistung $P$ an einem äußeren Lastwiderstand $R_a$, wenn eine Spannungsquelle mit Leerlaufspannung $U_0$ und Innenwiderstand $R_i$ vorliegt. Sie ergibt sich aus dem Ohm’schen Gesetz und der Leistungsformel $P = U^2 / R$, wobei die Klemmenspannung $U$ durch den Spannungsteiler von $R_I$ und $R$ bestimmt wird.

Der Leistungspunkt tritt bei $R_a = R_I$ auf. Aus dem Fit lassen sich somit $U_q$, $R_I$ und die maximale Leistung durch die Formel $P_\text{max}$ direkt bestimmen.
So haben sich folgende Werte ermitteln lassen:

Gefittete Quellspannung: $U_q = 2.786 \pm 0.046 V$
Gefitteter Innenwiderstand: $R_I = 86.0 \pm 2.663 Ω$
Maximale Leistung $P_\text{max} = 0.023 \pm 0.0000015696 W$ bei $R = 86.0 Ω$

```python
# | fig-cap: Leistung $P$ über den Außenwiderstand $R_a$ mit Curve-Fit über die gesammten Messwerte um Quellspannung und Innenwiederstand zu approximieren
# | label: fig-aufgabe-2-2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 1) Daten einlesen und P berechnen
df = pd.read_csv('./Daten/Aufgabe_1_Daten.csv')
df['I[mA]'] *= 1e-3    # mA → A
df['P[W]'] = df['U[V]'] * df['I[mA]']

# 2) Modell definieren: P(R) = a * R / (R + b)**2  (a = U0^2, b = R_i)
def P_model(R, a, b):
    return a * R / (R + b)**2

# 3) Curve-Fit durchführen
p0 = [ (df['U[V]'].max())**2, df['R_L[Ohm]'][df['P[W]'].idxmax()] ]  # Startwerte: a≈U_max^2, b≈R bei max P
popt, pcov = curve_fit(P_model,
                       df['R_L[Ohm]'].values,
                       df['P[W]'].values,
                       p0=p0)

a_fit, Ri_fit = popt
U0_fit = np.sqrt(a_fit)
Ropt   = Ri_fit
Popt   = P_model(Ropt, *popt)

#print(f"Gefittete Quellspannung Uq = {U0_fit:.3f} V")
#print(f"Gefitteter Innenwiderstand Ri = {Ri_fit:.1f} Ω")
# Berechnung der Abweichung für P_max
sigma_Popt = np.sqrt(
    (2 * Ropt / (Ropt + Ri_fit)**3 * sigma_U0)**2 + 
    (2 * U0_fit**2 * (Ropt - Ri_fit) / (Ropt + Ri_fit)**3 * sigma_Ri)**2
)

print(f"Maximale Leistung Pmax = {Popt:.3f} ± {sigma_Popt:.10f} W bei R = {Ropt:.1f} Ω")
# Berechnung der Abweichung
sigma_Ri = np.sqrt(pcov[1, 1])  # Standardabweichung für Ri
sigma_U0 = np.sqrt(pcov[0, 0]) / (2 * U0_fit)  # Standardabweichung für U0

#print(f"Abweichung der gefitteten Quellspannung σ_Uq = {sigma_U0:.3f} V")
#print(f"Abweichung des gefitteten Innenwiderstands σ_Ri = {sigma_Ri:.3f} Ω")

# 4) Plot inkl. Fit und Maximum
R_plot = np.logspace(np.log10(df['R_L[Ohm]'].min()),
                     np.log10(df['R_L[Ohm]'].max()), 500)
P_plot = P_model(R_plot, *popt)

plt.figure(figsize=(10,6))
plt.plot(df['R_L[Ohm]'], df['P[W]'], 'o', label='Messwerte')
plt.plot(R_plot, P_plot, '-', lw=2, label='Fit')
#plt.axvline(Ropt, color='orange', ls='--', label=f'$R_i$ = {Ropt:.1f} Ω → P$_max$')
#plt.scatter([Ropt], [Popt], color='orange', zorder=5)
plt.xscale('log')
plt.xlabel('$R_a$ [Ω]')
plt.ylabel('P [W]')
plt.title('Leistung $P$ über den Außenwiderstand $R_a$ mit Fit')
plt.legend()
plt.grid()
plt.show()
```
