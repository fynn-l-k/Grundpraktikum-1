## Ermittelung der Gitterkonstante
In der ersten Aufgabe ging es darum die Gitterkonstante des Reflexionsgitters zu ermitteln. Dabei nutzeten wir eine die grünen Spektrallinien, mit einer Wellenlänge von $\lambda = 546.1$ einer Hg-Spektrallampe. Ermittelt wurde $\phi$ sowie $\phi_0$.

Um nun auf die die Gitterkonstante zu kommen wird, die aus der Experimentier-Beschreibung die folgende Formel, der Phasendiffernez genommen.
$$\delta = 2 \pi \frac{g(sin \alpha - sin \beta)}{\lambda}$$
Dabei kann $\delta$ geschrieben werden als $\delta = 2k\pi$, wobei k die Beugungsordnung ist [Formel ist aus dem GP Buch]. So lässt sich die vorherige Formle umstellen nach $g$.
$$g = \frac{k \lambda}{sin \alpha - sin \beta}$$
Mit dieser Formel, kann man nun die Gitterkonstante für jede Messung ausrechnen, was in der folgenden Tabelle gemahct wurde.


| Beugungsordnung | $\alpha$ | $\beta$ | $\varphi$ | $\varphi_0$ | g - Gitterkonstnate |
|-----------------|----------|---------|-----------|-------------|---------------------|
| 0\.             | 59.955°  | 59.955° | 60.09°    | 0.4061°     | \-                  |
| 1\.             | 59.955°  | 30.775° | 89.27°    | 0.4061°     | 1.543               |
| 2\.             | 59.955°  | 8.985   | 111.06°   | 0.4061°     | 1.539               |
| 3\.             | 59.955°  | -11.475 | 131.52°   | 0.4061°     | 1.539               |
|                 |          |         |           |             | Durchschnitt        |
|                 |          |         |           | 0.4061°     | 1.540               |

: Tabelle mit allen bestimmten Winkeln, von der 0ten und 3ten Ordnung und die jeweilige Ermittelte Gitterkonstante


```python
# Gitterkonstante für alle Winkel
import numpy as np

winkel_pro_ordnungszahl = np.array([89.27, 111.06, 131.52]) #phi
erste_ordnung = 60.09 #phi_0

alpha = (180 - erste_ordnung)/2 #minus korre
_lambda = 546.1 * 10**(-9)
def g(winkel, k):
    return (k * _lambda)/(np.sin(alpha * np.pi/180)-np.sin(winkel * np.pi/180))

gitter_konst = np.array([])

for i in range(len(winkel_pro_ordnungszahl)):
    print(180 - winkel_pro_ordnungszahl[i] - alpha)
    gitter_konst = np.append(gitter_konst, g(180 - winkel_pro_ordnungszahl[i] - alpha, i + 1))

print(gitter_konst)
print(np.mean(np.abs(gitter_konst)))
```
