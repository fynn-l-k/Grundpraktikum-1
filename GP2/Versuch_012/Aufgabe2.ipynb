Für die zweite Aufgabe wurde die Hg-Spektrallampe ausgetauscht für eine Na-Spektrallampe, außerhalb davon blieb der Aufbau gleich. Es wurde wieder damit begonnen, den Korrekturwinkel $\phi_0$ zu bestimmen. Für diesen haben wir $\phi_0 = 359,5928°$ gemessen, oder auch $\phi_0 = -0,4072°$. Genau wie in Aufgabe 1 wurde auch das Reflexionsgitter so eingestellt, das bei ca. 60° das Maximum 0. Ordnung zu sehen ist. Genauer liegt dies bei $Max_0 = 60,1078°$.

Bei der Na-Spektrallampe kann man nicht nur einzelen Maxima sehen sondern durch die Spin-Bahn-Kopplung bei Natrium kann man 4 verschiedene Dubletts erkennen in den Farben Rot, Orange, Grün und Blau. Die Maxima erster Ordnung von allen Dubletts wurden gemessen und folgende Werte haben wir abgelesen:

Da bei $Max_0$ gilt $\alpha = \beta$, können wir wieder mit der Formel $\alpha = \frac{1}{2}(180° - (Max_0 - \phi_0))$ berechnen. Daraus folgt $\alpha = \frac{1}{2}(180° - (60,1078° + 0,4072))= 59,7425°$

Für die Berechnung der Wellenlänge nutzen wir die Formel: 
$$\lambda = \frac{g}{n}(sin(\beta_n) - sin(\alpha))$$
wobei wir immer nur das erste Maximum betrachten, weshalb immer n=1 gilt. Den Winkel für $\beta$ können wir berechnen mit $\beta = 180° - (\alpha + (\phi - \phi_0))$.

```python
import math

blau_1 = 87.27444444
blau_2 = 87.29222222
gruen_1 = 90.35611111
gruen_2 = 90.37666667
orange_1 = 91.23833333
orange_2 = 91.25777778
rot_1 = 92.35777778
rot_2 = 92.38000000
alpha = 59.7425
g = 1.54

def get_wavelength(color):
    beta = 180 - alpha - color
    wavelength = g * (math.sin(math.radians(beta)) - math.sin(math.radians(alpha)))
    return wavelength

def main():
    colors = {
        "blau_1": blau_1,
        "blau_2": blau_2,
        "gruen_1": gruen_1,
        "gruen_2": gruen_2,
        "orange_1": orange_1,
        "orange_2": orange_2,
        "rot_1": rot_1,
        "rot_2": rot_2,
    }

    for color_name, angle in colors.items():
        wavelength = get_wavelength(angle)
        print(f"{color_name}: {wavelength:.6f} nm")

if __name__ == "__main__":
    main()
```
