## Ermittelung der Gitterkonstante
In der ersten Aufgabe ging es darum die Gitterkonstante des Reflexionsgitters zu ermitteln. Dabei nutzeten wir eine die grünen Spektrallinien, mit einer Wellenlänge von $\lambda = 546.1$ nm einer Hg-Spektrallampe. Ermittelt wurde $\varphi$ sowie der Korrekturwinkel.

Um nun auf die die Gitterkonstante zu kommen wird, die aus der Beschreibung des Experimentes die folgende Formel, der Phasendifferenz genommen.
$$\delta = 2 \pi \frac{g(sin \alpha - sin \beta)}{\lambda}$$
Dabei kann $\delta$ geschrieben werden als $\delta = 2k\pi$, wobei $k$ die Beugungsordnung ist ^["Diese Formel wurde aus dem Buch "Physikalischen Praktikum", 14. Auflage, Hrsg. W. Schenk, F. Kremer, Optik, 2.3, S.263 entnommen"]. So lässt sich die vorherige Formel umstellen nach $g$.
$$g = \frac{k \lambda}{sin \alpha - sin \beta}$$
Mit dieser Formel, kann man nun die Gitterkonstante für jede Messung ausrechnen, was in der folgenden Tabelle gemacht wurde.

In der Tabelle wurde mit den Rohwerten und mit den Korrekturwinkel $\varphi_0$ die Gitterkonstante $g$ berechnet.

Es ergibt sich so aus der Tabelle eine durchschnittliche Gitterkonstante von $1.540 \cdot 10 ^{-6}m$ ohne Korrekturwinkel und eine durchschnittliche Gitterkonstante von $1.544 \cdot 10 ^{-6}$. Dabei gilt ein relativer Fehler mit der Fehlerfortpflanzung von 0.15%.


```python
# Gitterkonstante für alle Winkel
import numpy as np

korrektur_winkel = 0.4061

winkel_pro_ordnungszahl = np.array([89.27, 111.06, 131.52]) #phi
winkel_pro_ordnungszahl = winkel_pro_ordnungszahl - korrektur_winkel
erste_ordnung = 60.09 - korrektur_winkel #phi_0

alpha = (180 - erste_ordnung)/2 #minus korre
_lambda = 546.1 * 10**(-9)
def g(winkel, k):
    return (k * _lambda)/(np.sin(alpha * np.pi/180)-np.sin(winkel * np.pi/180))

gitter_konst = np.array([])

for i in range(len(winkel_pro_ordnungszahl)):
    print("BETA: ", 180 - winkel_pro_ordnungszahl[i] - alpha)
    gitter_konst = np.append(gitter_konst, g(180 - winkel_pro_ordnungszahl[i] - alpha, i + 1))

#print(gitter_konst)
print("Gemittelte Gitterkonstnate:")
print(np.mean(np.abs(gitter_konst)))
print("Berechnetes Alpha:")
print(alpha)

#print(winkel_pro_ordnungszahl)
```

# Aufgabe 2

Genau wie in Aufgabe 1 wurde auch das Reflexionsgitter so eingestellt, das bei ca. 60° das Maximum 0. Ordnung zu sehen ist. Genauer liegt dies bei $Max_0 = 60,1078°$.

Bei der Na-Spektrallampe kann man nicht nur einzelen Maxima sehen sondern durch die Spin-Bahn-Kopplung bei Natrium kann man 4 verschiedene Dubletts erkennen in den Farben Rot, Orange, Grün und Blau. Die Maxima erster Ordnung von allen Dubletts wurden gemessen und folgende Werte haben wir abgelesen:

Da bei $Max_0$ gilt $\alpha = \beta$, können wir wieder mit der Formel $\alpha = \frac{1}{2}(180° - (Max_0 - \phi_0))$ berechnen. Daraus folgt $\alpha = \frac{1}{2}(180° - (60,1078° + 0,4072))= 59,7425°$

Für die Berechnung der Wellenlänge nutzen wir die Formel: 
$$m \cdot \lambda = g \cdot (sin(\alpha) - sin(\beta))$$
wobei wir immer nur das erste Maximum betrachten, weshalb immer m=1 gilt. Den Winkel für $\beta$ können wir berechnen mit $\beta = 180° - (\alpha + (\phi - \phi_0))$.

Für die Gitterkonstante g nutzen wir den aus Aufgabe 1 berechneten Wert: $g = 1.544 \cdot 10 ^{-6}m$

```python
import math
import numpy as np

# Winkelwerte in Grad
blau_1 = 87.27444444
blau_2 = 87.29222222
gruen_1 = 90.35611111
gruen_2 = 90.37666667
orange_1 = 91.23833333
orange_2 = 91.25777778
rot_1 = 92.35777778
rot_2 = 92.38

korrekturwinkel = 0.4072
alpha = 59.7425
d = 1.544e-6  # Gitterkonstante in Metern

def get_wavelength(phi):
    beta = 180 - (alpha + (phi - korrekturwinkel))  # Richtige Formel!
    wavelength = d * (np.sin(alpha * np.pi/180) - (np.sin(beta * np.pi/180))) * 1e9
    return wavelength

def main():
    colors = {
        "blau_1": blau_1,
        "blau_2": blau_2,
        "gruen_1": gruen_1,
        "gruen_2": gruen_2,
        "orange_1": orange_1,
        "orange_2": orange_2,
        "rot_1": rot_1,
        "rot_2": rot_2
    }

    for name, angle in colors.items():
        wl = get_wavelength(angle)
        print(f"{name}: {wl:.3f} nm")

if __name__ == "__main__":
    main()
```

In Aufgabe 3 soll das Auflösungsvermögen des Gitters berechnet werden. Die Formel dafür kann aus der Experimentierbeschreibung entnommen werden.
$$A = \frac{\lambda}{\Delta \lambda}=kN$$
Wobei $k$ die Beugungsordnung und $N$ die Anzahl der beleuchteten Spalten ist. In der Aufgabestellung steht expilzit, dass, die wirksame Gitterbreite $b=50\text{mm}$ ($\pm 0.5%$) beträgt. Daraus lassen sich die Anzahl der Spalten ermitteln, indem man die wirksame Gitterbreite durch die Gitterkonstante mit Korrekturwinkel $g = 1.544 \cdot 10^{-6}$ teilt.
So ergibt sich die Formel der Auflösungsvermögen, welche wir nutzen mit, $A = k \frac{b}{g}$.

## Fehlerbetrachtung und Diskussion
Mit den relativen Fehlern, den Größen, Gitterkosntante $g$ und Gitterbreite $b$ ergibt sich auch für das Auflösungsvermögen ein relativer Forpflanzungsfehler.

$$\frac{\delta A}{A} = \frac{\delta N}{N} = \sqrt{\left( \frac{\delta g}{g} \right)^2\left( \frac{\delta b}{b} \right)^2} = \sqrt{0.005^2+0.0015^2} \approx 5.22 \cdot 10^{-3}$$

Dieser wurde benutzt um in der Tabelle den absoluten Fehler für den Wert anzugeben. Dieser ist klein gegenüber dem Auflösungsvermögen.
