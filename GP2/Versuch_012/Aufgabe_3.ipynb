# Aufgabe 3
In Aufgabe 3 soll das Auflösungsvermögen des Gitters berechnet werden. Die Formel dafür kann aus der Experimentierbeschreibung entnommen werden.
$$A = \frac{\lambda}{\Delta \lambda}=kN$$
Wobei $k$ die Beugungsordnung und $N$ die Anzahl der beleuchteten Spalten ist. In der Aufgabestellung steht expilzit, dass, die wirksame Gitterbreite $b=50\text{mm}$ ($\pm 0.5%$) beträgt. Daraus lassen sich die Anzahl der Spalten ermitteln, indem man die wirksame Gitterbreite durch die Gitterkonstante $g = 1.540 \cdot 10^{-6}$ teilt.
So ergibt sich die Formel der Auflösungsvermögen, welche wir nutzen mit, $A = k \frac{b}{g}$.


| Beugungsordnung $k$ | Auflösungsvermögen $A$ |
|---------------------|------------------------|
| $k = 1$             | 32467.532              |
| $k = 2$             | 64935.065              |
| $k = 3$             | 97402.597              |

: AUflösungsvermögen zu verschiedenen Beugungsordnungen $k = 1, 2, 3$
