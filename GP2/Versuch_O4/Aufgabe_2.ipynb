## Herleitung der Gleichung zum Bestimmen der Brennweiten $f_1$ und $f_2$
Zur Bestimmung der Brennweiten $f_1$ und $f_2$ wurde, die Besselgleichung genommen. Da man dort die Variablen Bildabstand $b$ und Gestandsabstand $g$ nicht benötigt. Die Besselgleichung lautet wie folgt:

$$f = \frac{s² - e²}{4s}$$

Dabei ist $s=g+b$ der feste Abstand zwischen dem Schirm und dem Gegenstand und $e$ der Abstand an dem die Linsen ihren Brennpunkt auf dem Schrim haben und ein scharfes Bild darauf zu erkennen ist.

Mit der folgenden Formel können wir dann, die Brennweite $f_{sys}$ für dass gesammt System erhalten. Dabei ist $f_n$ die Brennweite der $n$-ten Linse und $d_l$ der Abstand zwischen den beiden Linsen.

$$\frac{1}{f_{sys}} = \frac{1}{f_1} + \frac{1}{f_2} - \frac{d_l}{f_1 f_2}$$
