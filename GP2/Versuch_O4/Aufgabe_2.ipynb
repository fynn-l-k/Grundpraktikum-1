## Positionen der 2f-2f Konfiguration und Einstellungen der Kamera

In Aufgabe 2 galt es, mithilfe der in 1 errechneten Brennweiten das System in die 2f-2f Konfiguration zu bringen. In dieser Konfiguration wissen wir, dass auf dem Schirm das Bild genau invertiert aber in der gleichen größe erscheint. Somit können wir den Maßstab der Kamera einstellen und zuweisen, wie die skalierung von pixel zu mm ist.

Folgende Positionen hatten unsere Bauteile in der 2f-2f Konfiguration:
|Linse|Position Kamera (mit Korrektur) [mm]| Position Linse [mm]| Position Kamera [mm] |
|--------:|------------:|------:|----------:|
|1|40 $\implies$ 70|352.8|635.6|
|1|40 $\implies$ 70|539.63|1009.27|

Für die Einstellung haben wir Linse 2 genommen und einem Milimeter die richtige größe zugeordnet. Mit der ersten Linse haben wir dies dann nochmal überprüft und es war dann noch $0,98$ mm, was eine normale Abweichung ist, da wir nur das Linear unten als Einstellungsgröße von den Geräten hatten und somit leichte Abweichungen vorkommen können. Somit hatten wir eine Skalierung der Kamera von: $Scale = 352,3361 \frac{pixels}{mm}$.
