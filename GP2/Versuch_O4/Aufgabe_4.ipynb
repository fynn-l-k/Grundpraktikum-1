## Berechnung der Systembrennweite


Mit der folgenden Formel können wir dann, die Brennweite $f_{sys}$ für dass gesammt System erhalten. Dabei ist $f_n$ die Brennweite der $n$-ten Linse und $d_l$ der Abstand zwischen den beiden Linsen.

$$\frac{1}{f_{sys}} = \frac{1}{f_1} + \frac{1}{f_2} - \frac{d_l}{f_1 f_2}$$

So erhalten wir eine theoretische Systembrennweite von $f_{sys,t} = 205.95$ mm.

## Vergleich mit Systembrennweite aus Aufgabe 1
Bei Aufgabe 1 erhalten wir die Systembrennweiten $f_{sys}(g') = 200.631 \pm 2.017$ mm und $f_{sys}(b') = 203.295 \pm 1.109$ mm, diese beiden Werte sind sehr nah, selbst mit den Unsicherheiten, an unserem theoretischen Wert für die Systembrennweite von $f_{sys,t} = 205.95$ mm. 

Die kleine Abweichung vom theoretischen Ergbniss, zeigt dass wir sehr gute Messwerte haben. Was vermutlich durch die quantitativität der Messpunkte möglich ist. Da durch kann der Fit sich noch besser anpassen und die leichten Schwankungen am anfang so besser ausbessern.
Außerdem können die vielen Messunsicherheiten auch zu dieser Abweichung führe. So wurden die Abstände immer auf einem Lineal auf der optischen Bank, abgelesen ($\pm 0.5$ mm). Zudem wurde in der Software, zwar versucht genau auf dem Punkt zu sein, bei der Skalierungsmessung. Jedoch hat dass mit dem TOuchpad des Laptops nicht immer einfach funktioniert, selbst beim besten willen, weshalb dort bestimmt eine Abweichung von $\pm 0.05$ mm dazugekommen ist. Zudem es außerdem ein optischer Versuch waren können auch Optische Fehler die Systembrennweite abweichen lassen, da der Experimentieraufbau nicht theoretissch ideal war. Es könnte zu sphärischer, chromischer Abberation oder zu Antistagmatismus gekommen sein, die die Kamera aufgefasst hat und welche die Werte leicht verändert.
