# Aufgabe 3
In der dritten Aufgabe sollte eine maßstaabsgetreue Skizze gezeichnet werden, welche unserm Versuchsaufbau entspricht für den dritten Versuch. Für die SKizze wurde die theoretische Brennweite benutzt um Fehlerforpflanzung zu mindern.

![Maßstaabsgetreue Skizze, mit der Entstehung des Bildes. Dabei entspricht 1 Kästchen auf der y-Achse 0.3mm und ein Kästchen auf der x-Achse 100mm](./Skizze.jpeg)

In der Skizze erhalten wir ein invertiertes Bild, dies war aber nicht im Versuch so. Wenn wir dass Bild mit den Linsen ermitteln, kommenn wir, mit dem Zwischenbild auf ein nicht invertiertes Bild. Die Konstruktion mit den Hauptebenen ist aber einfacher und da in der Aufgabenstellungen keine genaue Konstruktionsart verlangt wurde. Nur dass man die Bildgröße aus der Konstruktions ermitteln kann.
Die Bildorientierung ist dabei für die Auswertung der experimentellen Daten irrelevant, da sie nicht die Brennweite, Bildposition oder Vergrößerung beeinflusst. Beide Konstruktionsmethoden liefern die gleichen quantitativen Ergebnisse für diese Parameter.

Aus der Skizze lässt sich mit dem Lineal $0.25$ mm ablesen, wenn man dies vergleicht mit dem gemessenen Wert von $0.246$ mm, fällt auf dass der Wert sehr nah bei dem experimentellen Wert liegt. Dass zeigt auch nochmal dass die Konstruktionsart nicht relavent ist. Der Wert liegt auch trotz SKizzen und Messfehler mit Lineal nur $0.034$ mm, was eine sehr kleine und gute Abweichung ist.
