```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df = pd.read_csv('Messdaten/Aufgabe3.csv')
df.columns = df.columns.str.replace('"', '').str.strip()



# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df['Skalierung'], (df['Pos_Kamera [mm]'])-df['Pos_Sys [mm]'], color="blue")
plt.ylabel("Pos_Kamera - Pos_Sys [mm]")
plt.xlabel("Skalierung")
plt.title("Pos_Kamera - Pos_Sys vs Skalierung")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
