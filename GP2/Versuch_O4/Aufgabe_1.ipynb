Die Werte für $b'$, $g'$ und $\gamma$ haben wir aus folgenden Gleichungen bekommen:
$$
g' = x_{systen} - x_{lampe} ;b' = x_{kamera} - x_{system} ; \gamma = \frac{B}{G} = \frac{Skalierung}{0,3}.
$$
Jetzt bauen wir eine Formel um aus dem Fit unserer Messwerte die Systembrennweite und die Hauptebenen zu bestimmen. Als erstes wissen wir das gilt:
$$
g' = g + h ; b' = b + h'
$$
Wir können eine Formel für $g$ und $b$ aus der Gleichung eines Linsensystems formen mit $\gamma$ = $b/g$:
$$
\frac{1}{f_{sys}} = \frac{1}{g} + \frac{1}{b} \implies \frac{1}{f_{sys}} = \frac{1}{g} + \frac{1}{\gamma \cdot g} = \frac{1}{g}\cdot \left( 1+ \frac{1}{\gamma} \right)
$$

Daraus:
$$
g = f_{sys}\cdot \left(1 + \frac{1}{\gamma} \right) \text{und } b = f_{sys}\cdot \left(1 +\gamma \right)
$$

Mit unserem wissen von vorher können wir die folgenden Fitfunktionen nutzen:
$$
g' = f_{sys} \cdot \frac{1}{\gamma} + f_{sys} + h \text{ und } b' = f_{sys} \cdot \gamma + f_{sys} + h'
$$

Also ist bei unseren linearen Fits $f_{sys}$ die Steigung und $f_{sys} + h$ oder $f_{sys} + h'$ der Achsenabschnitt, also wir können Hauptebene und Brennweite aus den Fits bestimmen.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# 0. Einstellungen
# ============================================================

# Position der Lampe in mm
LAMP_POS = 70.0  # mm

# Reale Länge der gemessenen Strecke auf der Skala (G) in mm
G_REAL = 0.3  # mm

# Dateiname deiner CSV
CSV_FILE = "Messdaten/Aufgabe3.csv"  # <- hier ggf. anpassen

# ============================================================
# 1. CSV einlesen
# ============================================================

df = pd.read_csv(CSV_FILE)

# Spalten auslesen (Namen müssen genau so heißen wie in deiner Tabelle)
pos_sys = df["Pos_Sys [mm]"].to_numpy()
pos_cam = df["Pos_Kamera [mm]"].to_numpy()
scale = df["Skalierung"].to_numpy()

# Sicherheitscheck
if not (len(pos_sys) == len(pos_cam) == len(scale)):
    raise ValueError("Spaltenlängen passen nicht zusammen!")

# ============================================================
# 2. g', b' und γ berechnen
# ============================================================

# g' = Abstand Linsensystem - Lampe
g_prime = pos_sys - LAMP_POS   # in mm

# b' = Abstand Kamera - Linsensystem
b_prime = pos_cam - pos_sys    # in mm

# Bildgröße B (proportional zur gemessenen Länge)
B = scale

# Abbildungsmaßstab γ = B / G
gamma = B / G_REAL
inv_gamma = 1.0 / gamma

# ============================================================
# 3. Fit für b'(γ): b' = f_sys * γ + (f_sys + h')
# ============================================================

coeff_b, cov_b = np.polyfit(gamma, b_prime, deg=1, cov=True)
m_b, c_b = coeff_b
err_m_b, err_c_b = np.sqrt(np.diag(cov_b))

f_sys_from_b = m_b
h_prime = c_b - f_sys_from_b

# ============================================================
# 4. Fit für g'(1/γ): g' = f_sys * (1/γ) + (f_sys + h)
# ============================================================

coeff_g, cov_g = np.polyfit(inv_gamma, g_prime, deg=1, cov=True)
m_g, c_g = coeff_g
err_m_g, err_c_g = np.sqrt(np.diag(cov_g))

f_sys_from_g = m_g
h = c_g - f_sys_from_g

# ============================================================
# 5. Ergebnisse ausgeben
# ============================================================
print("=== Fit aus b'(γ) ===")
print(f"f_sys (aus b') = {f_sys_from_b:.3f} ± {err_m_b:.3f} mm")
print(f"h'              = {h_prime:.3f} ± {err_c_b:.3f} mm")

print("\n=== Fit aus g'(1/γ) ===")
print(f"f_sys (aus g') = {f_sys_from_g:.3f} ± {err_m_g:.3f} mm")
print(f"h               = {h:.3f} ± {err_c_g:.3f} mm")

# ============================================================
# 6. Plots erstellen
# ============================================================

# Plot 1: b' vs gamma
gamma_fit = np.linspace(min(gamma)*0.9, max(gamma)*1.1, 200)
b_fit = m_b * gamma_fit + c_b

plt.figure()
plt.scatter(gamma, b_prime, label="Messwerte b'")
plt.plot(gamma_fit, b_fit, label="Fit: b' = m·γ + c")
plt.xlabel("γ = B / G")
plt.ylabel("b' (mm)")
plt.title("b' als Funktion von γ")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Plot 2: g' vs 1/gamma
inv_gamma_fit = np.linspace(min(inv_gamma)*0.9, max(inv_gamma)*1.1, 200)
g_fit = m_g * inv_gamma_fit + c_g

plt.figure()
plt.scatter(inv_gamma, g_prime, label="Messwerte g'")
plt.plot(inv_gamma_fit, g_fit, label="Fit: g' = m·(1/γ) + c")
plt.xlabel("1 / γ")
plt.ylabel("g' (mm)")
plt.title("g' als Funktion von 1/γ")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
