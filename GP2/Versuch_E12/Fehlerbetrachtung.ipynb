# Fehlerbetrachtung
Mögliche Fehlerquellen, sind die Messtolleranzen und Einstellungsfeinheiten, bei zum Beispiel der Frequenz, des Multimetergerät sowie des Picoscopes. Zudem wurde, für jeden Versuch ein idealer Schaltkreis angenommen, was in der echten Welt natürlich nicht der Fall ist. So hat jedes elektrisches Bauteil einen Eigenwiederstand der nicht beachtet werden konnte.
