# Aufgabe 0

Im Rahmen dieser Vorbereitung wird die Funktionsweise eines
$\lambda/4$-Plättchens sowie die Intensität nach einem Analysator für
verschiedene Polarisationszustände berechnet. Der Polarisationszustand
wird durch Jones-Vektoren $\vec{J}$ und optische Komponenten durch
Jones-Matrizen $M$ beschrieben.

## Grundlagen

Für linear polarisiertes Licht im Winkel $\alpha$ zur Vertikalen gilt:

$$\vec{J}_\alpha = \begin{pmatrix} \sin \alpha \\ \cos \alpha \end{pmatrix}$$

Für rechts- ($+$) bzw. linkszirkular ($-$) polarisiertes Licht gilt:

$$\vec{J}_+ = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ -i \end{pmatrix}, \quad \vec{J}_- = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ i \end{pmatrix}$$

Die Jones-Matrizen für einen Analysator (Winkel $\theta$) und eine
$\lambda/4$-Platte (Winkel $\varphi$) sind:

$$M_\theta = \begin{pmatrix} \sin^2 \theta & \sin \theta \cos \theta \\ \sin \theta \cos \theta & \cos^2 \theta \end{pmatrix}, \quad 
M_{\lambda/4}^\varphi = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 + i \cos(2\varphi) & -i \sin(2\varphi) \\ -i \sin(2\varphi) & 1 - i \cos(2\varphi) \end{pmatrix}$$

Die Intensität nach dem System berechnet sich zu

$I = I_0 |M_n \cdot \dots \cdot M_1 \cdot \vec{J}|^2$.

## Berechnung der Intensitätsfunktionen

### (a) Linear polarisiertes Licht & Analysator

Die Intensität folgt dem Gesetz von Malus:

$$I(\alpha) = I_0 |M_\theta \vec{J}_\alpha|^2 = I_0 \cos^2(\alpha - \theta)$$

### (b) Zirkular polarisiertes Licht & Analysator

Da zirkulares Licht keine Vorzugsrichtung hat, ist die Intensität nach
dem Analysator konstant:

$$I_\pm = I_0 |M_\theta \vec{J}_\pm|^2 = \frac{I_0}{2}$$

### (c) Linear polarisiertes Licht, $\lambda/4$-Platte und Analysator

Hier ergibt die Verknüpfung der Matrizen:

$$I = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_\alpha|^2 = \frac{I_0}{2} \left[ \cos^2(\alpha + \theta - 2\varphi) + \cos^2(\alpha - \theta) \right]$$

### (d) Zirkular polarisiertes Licht, $\lambda/4$-Platte und Analysator

Durch das $\lambda/4$-Plättchen wird zirkulares Licht wieder linear polarisiert. Die Intensität hängt vom Winkel des Analysators zum Plättchen ab:

$$I_+ = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_+|^2 = \frac{I_0}{2} (1 - \sin(2\theta - 2\varphi))$$
$$I_- = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_-|^2 = \frac{I_0}{2} (1 + \sin(2\theta - 2\varphi))$$
