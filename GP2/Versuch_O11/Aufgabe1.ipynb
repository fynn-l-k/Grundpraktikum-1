```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df1 = pd.read_csv('messdaten/Aufgabe1_0.csv')
df1.columns = df1.columns.str.replace('"', '').str.strip()

df2 = pd.read_csv('messdaten/Aufgabe1_30.csv')
df2.columns = df2.columns.str.replace('"', '').str.strip()

df3 = pd.read_csv('messdaten/Aufgabe1_45.csv')
df3.columns = df3.columns.str.replace('"', '').str.strip()


# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df1['Winkel[alpha]'], df1['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke", color="red", marker='o', markersize=4, linewidth=0)
plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke", color="green", marker='o', markersize=4, linewidth=0)
plt.xlabel("Winkel [alpha]")
plt.ylabel("Intensität [V]")
plt.title("Lichtstärke über den Winkel")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()
```
