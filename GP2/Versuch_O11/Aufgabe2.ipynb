# Aufgabe 2

Für die zweite Aufgabe hatten wir 3 Blackboxen, bei denen wir nicht wussten, wie das Licht polarisiert ist. Die Aufgabe war es genau diese Polarisation durch messen in verschiedenen Aufbauten herauszufinden. Dafür haben wir uns an das folgende Schema aus dem Skript gehalten:
## Schema zur Bestimmung der Polarisation (mit Analysator und ggf. $\lambda$/4-Plättchen)

### 1) Keine Intensitätsvariation beim Drehen des Analysators
→ Bringe ein $\lambda$/4-Plättchen vor dem Analysator ein (in beliebiger Orientierung) und beobachte erneut.

(a) Weiterhin keine Intensitätsvariation beim Drehen des Analysators
→ natürlich unpolarisiertes Licht

(b) Es gibt ein Intensitätsminimum bei einer Analysatorstellung
- i) Intensität kann in einer Stellung ganz verschwinden
  → zirkular polarisiertes Licht
- ii) Intensität verschwindet nie vollständig (für alle Analysatorstellungen bleibt Restintensität)
  → Überlagerung aus zirkular polarisiertem und unpolarisiertem Licht

### 2) Intensitätsvariation beim Drehen des Analysators
(a) Intensität kann in einer Analysatorstellung ganz verschwinden
→ linear polarisiertes Licht

(b) Intensität verschwindet nie vollständig (Restintensität für alle Analysatorstellungen)
→ Bringe ein $\lambda$/4-Plättchen vor dem Analysator ein, so dass die schnelle Achse des Plättchens
parallel zur Analysatorstellung bei maximaler oder minimaler Intensität liegt.

- i) Intensität kann dann in einer Stellung ganz verschwinden
  → elliptisch polarisiertes Licht

- ii) Intensität verschwindet weiterhin nie vollständig
  - A) Minimale Intensität bei derselben Analysatorstellung wie ohne $\lambda$/4-Plättchen
    → Überlagerung von linear polarisiertem und unpolarisiertem Licht
  - B) Minimale Intensität bei einer anderen Analysatorstellung als ohne $\lambda$/4-Plättchen
    → Überlagerung von elliptisch polarisiertem und unpolarisiertem Licht



## Blackbox 1

Wir haben bei der Blackbox 1 angefangen die Intensität zu messen ohne das $\lambda$/4 Plättchen. Dabei haben wir den analysator immer um 10° gedreht und somit einen Winkel-Intensität graphen bekommen. Bei diesem konnten wir feststellen, dass es eine intensitätsvariation beim drehen gab, die Intensität aber niemals auf 0 fällt. Das heißt wir müssen eine zweite Messung durchführen bei der wir die schnelle Achse des $\lambda$/4 Plättchens parallel zur minimalen Intensität stellen. In unserer Messung haben wir das Minimum bei 7° gefunden und deshalb die schnelle Achse auch auf 7° gestellt. Die zweite Messung kann man auch unten im Graphen sehen und man sieht, dass die Intensität weiterhin nie vollständig verschwindet und das minimum sich fast nicht verschoben hat. Die verschiebung schieben wir auf Messungenauigkeiten und da die einzelnen Winkel per Hand eingestellt werden kann es auch dort zu kleineren unterschieden kommen. Daher nehmen wir für die Blackbox 1 an, dass es sich um eine Überlagerung von linear polarisiertem und unpolarisiertem Licht handelt.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df1 = pd.read_csv('messdaten/Aufgabe2_BB1_1.csv')
df1.columns = df1.columns.str.replace('"', '').str.strip()

df2 = pd.read_csv('messdaten/Aufgabe2_BB1_2.csv')
df2.columns = df2.columns.str.replace('"', '').str.strip()

# df3 = pd.read_csv('messdaten/Aufgabe1_45.csv')
# df3.columns = df3.columns.str.replace('"', '').str.strip()


# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df1['Winkel[alpha]'], df1['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=1)
plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke", color="red", marker='o', markersize=4, linewidth=1)
# plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke", color="green", marker='o', markersize=4, linewidth=1)
plt.xlabel("Winkel [alpha]")
plt.ylabel("Intensität [V]")
plt.title("Lichtstärke über den Winkel")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()
```
