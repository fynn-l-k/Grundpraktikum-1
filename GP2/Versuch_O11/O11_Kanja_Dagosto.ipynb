# Aufgabe 0

Im Rahmen dieser Vorbereitung wird die Funktionsweise eines $\lambda/4$-Plättchens sowie die Intensität nach einem Analysator für verschiedene Polarisationszustände berechnet. Der Polarisationszustand wird durch Jones-Vektoren $\vec{J}$ und optische Komponenten durch Jones-Matrizen $M$ beschrieben.

## Grundlagen

Für linear polarisiertes Licht im Winkel $\alpha$ zur Vertikalen gilt:

$$\vec{J}_\alpha = \begin{pmatrix} \sin \alpha \\ \cos \alpha \end{pmatrix}$$

Für rechts- ($+$) bzw. linkszirkular ($-$) polarisiertes Licht gilt:

$$\vec{J}_+ = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ -i \end{pmatrix}, \quad \vec{J}_- = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ i \end{pmatrix}$$

Die Jones-Matrizen für einen Analysator (Winkel $\theta$) und eine
$\lambda/4$-Platte (Winkel $\varphi$) sind:

$$M_\theta = \begin{pmatrix} \sin^2 \theta & \sin \theta \cos \theta \\ \sin \theta \cos \theta & \cos^2 \theta \end{pmatrix}, \quad 
M_{\lambda/4}^\varphi = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 + i \cos(2\varphi) & -i \sin(2\varphi) \\ -i \sin(2\varphi) & 1 - i \cos(2\varphi) \end{pmatrix}$$

Die Intensität nach dem System berechnet sich zu

$I = I_0 |M_n \cdot \dots \cdot M_1 \cdot \vec{J}|^2$.

## Berechnung der Intensitätsfunktionen

### Linear polarisiertes Licht & Analysator

Die Intensität folgt dem Gesetz von Malus:

$$I(\alpha) = I_0 |M_\theta \vec{J}_\alpha|^2 = I_0 \cos^2(\alpha - \theta)$$

### Zirkular polarisiertes Licht & Analysator

Da zirkulares Licht keine Vorzugsrichtung hat, ist die Intensität nach
dem Analysator konstant:

$$I_\pm = I_0 |M_\theta \vec{J}_\pm|^2 = \frac{I_0}{2}$$

### Linear polarisiertes Licht, $\lambda/4$-Platte und Analysator

Hier ergibt die Verknüpfung der Matrizen:

$$I = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_\alpha|^2 = \frac{I_0}{2} \left[ \cos^2(\alpha + \theta - 2\varphi) + \cos^2(\alpha - \theta) \right]$$

### Zirkular polarisiertes Licht, $\lambda/4$-Platte und Analysator

Durch das $\lambda/4$-Plättchen wird zirkulares Licht wieder linear polarisiert. Die Intensität hängt vom Winkel des Analysators zum Plättchen ab:

$$I_+ = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_+|^2 = \frac{I_0}{2} (1 - \sin(2\theta - 2\varphi))$$
$$I_- = I_0 |M_\theta M_{\lambda/4}^\varphi \vec{J}_-|^2 = \frac{I_0}{2} (1 + \sin(2\theta - 2\varphi))$$

# Aufgabe 1

In dieser Aufgabe wird die Intensität von linear polarisiertem Licht ($\alpha \approx 0°$) nach dem Durchlaufen einer $\lambda/4$-Platte (Winkel $\varphi$) und eines Analysators (Winkel $\theta$) untersucht. Die Messung der Intensität erfolgt über die Photodiodenspannung, da $U \~ I$. Für die Winkel $\varphi = 0°, 30°, 45°$ wurden Intensitätsprofile über $\theta$ aufgenommen und mittels der vorherigen Gleichung aus Aufgabe 0 gefittet, um die Parameter $\alpha$, $\phi$ und $I_0$ samt Unsicherheiten zu bestimmen. DIe Gleichung lautet wie folg:

$$I  = \frac{I_0}{2} \left[ \cos^2(\alpha + \theta - 2\varphi) + \cos^2(\alpha - \theta) \right]$$

```python
# | fig-cap: Lichtstärke in $V$ über den Winkel $\\theta°$ bei Verschiedenen $\varphi°$ vom $\lambda / 2 \text{- Plätchen}$

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV einlesen
df1 = pd.read_csv('messdaten/Aufgabe1_0.csv')
df1.columns = df1.columns.str.replace('"', '').str.strip()

df2 = pd.read_csv('messdaten/Aufgabe1_30.csv')
df2.columns = df2.columns.str.replace('"', '').str.strip()

df3 = pd.read_csv('messdaten/Aufgabe1_45.csv')
df3.columns = df3.columns.str.replace('"', '').str.strip()


# Plot für a_x
plt.figure(figsize=(10, 4))
plt.plot(df1['Winkel[alpha]'], df1['Intensität[V]'], label="Lichtstärke bei $\\varphi = 0°$", color="blue", marker='o', markersize=4, linewidth=1)
plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke bei $\\varphi = 30°$", color="red", marker='o', markersize=4, linewidth=1)
plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke bei $\\varphi = 45°$", color="green", marker='o', markersize=4, linewidth=1)
plt.xlabel("Winkel [$\\theta°$]")
plt.ylabel("Intensität [$V$]")
plt.title("Lichtstärke über den Winkel $\\theta°$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()


# plt.figure(figsize=(10, 4))
# plt.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], label="Lichtstärke", color="blue", marker='o', markersize=4, linewidth=0)
# plt.xlabel("Winkel [alpha]")
# plt.ylabel("Intensität [V]")
# plt.title("Lichtstärke über den Winkel")
# plt.grid(True)
# plt.legend()
# plt.tight_layout()
# plt.show()
```

```python
from scipy.optimize import curve_fit

def func0c(theta, phi, alpha, constant):
  #theta, phi, alpha = theta_phi_alpha
  return constant*0.5*((np.cos(np.deg2rad(alpha + theta - 2*phi)))**2 + (np.cos(np.deg2rad(alpha - theta))**2))


popt1_0, pcov1_0 = curve_fit(func0c, df1['Winkel[alpha]'], df1['Intensität[V]'], [0, 0, 0])

fig, ax = plt.subplots(figsize=(10, 4))
theta_fit = np.linspace(df1['Winkel[alpha]'].min(), df1['Winkel[alpha]'].max(), 200)
ax.plot(df1['Winkel[alpha]'], df1['Intensität[V]'], 'o', label='Messdaten', markersize=4)
ax.plot(theta_fit, func0c(theta_fit, *popt1_0), '-', label='Curve-Fit', linewidth=2)
ax.set_xlabel("Winkel [$\\theta°$]")
ax.set_ylabel("Intensität [$V$]")
ax.grid(True)
ax.legend()
ax.set_title("Intensität bei $\\varphi = 0°$")
plt.tight_layout()
plt.show()

print(f"Fitparameter für phi = 0°:")
print(f"  phi = {popt1_0[0]:.4f}° \\pm {np.sqrt(pcov1_0[0, 0]):.4f}°")
print(f"  alpha = {popt1_0[1]:.4f}° \\pm {np.sqrt(pcov1_0[1, 1]):.4f}°")
print(f"  I_0 = {popt1_0[2]:.4f} V \\pm {np.sqrt(pcov1_0[2, 2]):.4f} V")
```

```python
popt1_0, pcov1_0 = curve_fit(func0c, df2['Winkel[alpha]'], df2['Intensität[V]'], [0, 0, 0])

fig, ax = plt.subplots(figsize=(10, 4))
theta_fit = np.linspace(df2['Winkel[alpha]'].min(), df2['Winkel[alpha]'].max(), 200)
ax.plot(df2['Winkel[alpha]'], df2['Intensität[V]'], 'o', label='Messdaten', markersize=4)
ax.plot(theta_fit, func0c(theta_fit, *popt1_0), '-', label='Curve-Fit', linewidth=2)
ax.set_xlabel("Winkel [$\\theta°$]")
ax.set_ylabel("Intensität [$V$]")
ax.grid(True)
ax.legend()
ax.set_title("Intensität bei $\\varphi = 30°$")
plt.tight_layout()
plt.show()

print(f"Fitparameter für phi = 30°:")
print(f"  phi = {popt1_0[0]:.4f}° \\pm {np.sqrt(pcov1_0[0, 0]):.4f}°")
print(f"  alpha = {popt1_0[1]:.4f}° \\pm {np.sqrt(pcov1_0[1, 1]):.4f}°")
print(f"  I_0 = {popt1_0[2]:.4f} V \\pm {np.sqrt(pcov1_0[2, 2]):.4f} V")
```

```python
popt1_0, pcov1_0 = curve_fit(func0c, df3['Winkel[alpha]'], df3['Intensität[V]'], [0, 0, 0])

fig, ax = plt.subplots(figsize=(10, 4))
theta_fit = np.linspace(df3['Winkel[alpha]'].min(), df3['Winkel[alpha]'].max(), 200)
ax.plot(df3['Winkel[alpha]'], df3['Intensität[V]'], 'o', label='Messdaten', markersize=4)
ax.plot(theta_fit, func0c(theta_fit, *popt1_0), '-', label='Curve-Fit', linewidth=2)
ax.set_xlabel("Winkel [$\\theta°$]")
ax.set_ylabel("Intensität [$V$]")
ax.grid(True)
ax.legend()
ax.set_title("Intensität bei $\\varphi = 45°$")
plt.tight_layout()
plt.show()

print(f"Fitparameter für phi = 45°:")
print(f"  phi = {popt1_0[0]:.4f}° \\pm {np.sqrt(pcov1_0[0, 0]):.4f}°")
print(f"  alpha = {popt1_0[1]:.4f}° \\pm {np.sqrt(pcov1_0[1, 1]):.4f}°")
print(f"  I_0 = {popt1_0[2]:.4f} V \\pm {np.sqrt(pcov1_0[2, 2]):.4f} V")
```

## Analyse der Fit-Ergebnisse

Aus den durchgeführten Fits lassen sich folgende Erkenntnisse gewinnen:

Für $\varphi = 0°$:
Die Fitparameter zeigen Abweichungen von den erwarteten theoretischen Werten ($\alpha = 0°$, $\varphi = 0°$). Die großen Unsicherheiten in den Fitparametern deuten auf Probleme bei der Parameterbestimmung hin.

Für $\varphi = 30°$:
Ähnlich wie bei $\varphi = 0°$ zeigen sich erhebliche Abweichungen und große Fehlerbalken, was auf eine schlechte Konditionierung des Fits hindeutet.

Für $\varphi = 45°$:
Wie erwartet, ergibt sich ein nahezu konstantes Intensitätsprofil ($I(\theta) \approx I_0/2 = \text{const.}$). Dies bestätigt, dass bei $\varphi = 45°$ und $\alpha = 0° \text{bzw.} 90°$ die beiden Polarisationskomponenten gleich stark zur Gesamtintensität beitragen. Die leichten Schwankungen deuten auf kleine Fehleinstellungen hin.

Die theoretische Vorhersage für $\varphi = 45°$ wird experimentell bestätigt. Bei $\varphi = 0°$ und $\varphi = 30°$ zeigen die Daten das erwartete sinusförmige Verhalten, jedoch mit Abweichungen, die auf die Qualität unserer Werte zurückzuführen sind.

## Fehlerbetrachtung

Die Hauptfehlerquellen liegen in der mechanischen Justage der Optiken, dem Detektorrauschen sowie der Instabilität der Lichtquelle. Zudem erschwert die starke Parameterkopplung von $\varphi$ und $\alpha$ eine präzise Bestimmung. Besonders bei $\alpha$ übersteigt der Fit-Fehler den Parameterwert, was auf eine mangelnde Robustheit des Fitting-Verfahrens hindeutet.



# Aufgabe 2

Für die zweite Aufgabe hatten wir 3 Blackboxen, bei denen wir nicht wussten, wie das Licht polarisiert ist. Die Aufgabe war es genau diese Polarisation durch messen in verschiedenen Aufbauten herauszufinden. Dafür haben wir uns an das folgende Schema aus dem Skript gehalten:
## Schema zur Bestimmung der Polarisation (mit Analysator und ggf. $\lambda$/4-Plättchen)

### Keine Intensitätsvariation beim Drehen des Analysators
→ Bringe ein $\lambda$/4-Plättchen vor dem Analysator ein (in beliebiger Orientierung) und beobachte erneut.

(a) Weiterhin keine Intensitätsvariation beim Drehen des Analysators
→ natürlich unpolarisiertes Licht

(b) Es gibt ein Intensitätsminimum bei einer Analysatorstellung
- i) Intensität kann in einer Stellung ganz verschwinden
  → zirkular polarisiertes Licht
- ii) Intensität verschwindet nie vollständig (für alle Analysatorstellungen bleibt Restintensität)
  → Überlagerung aus zirkular polarisiertem und unpolarisiertem Licht

### Intensitätsvariation beim Drehen des Analysators
(a) Intensität kann in einer Analysatorstellung ganz verschwinden
→ linear polarisiertes Licht

(b) Intensität verschwindet nie vollständig (Restintensität für alle Analysatorstellungen)
→ Bringe ein $\lambda$/4-Plättchen vor dem Analysator ein, so dass die schnelle Achse des Plättchens
parallel zur Analysatorstellung bei maximaler oder minimaler Intensität liegt.

- i) Intensität kann dann in einer Stellung ganz verschwinden
  → elliptisch polarisiertes Licht

- ii) Intensität verschwindet weiterhin nie vollständig
  - A) Minimale Intensität bei derselben Analysatorstellung wie ohne $\lambda$/4-Plättchen
    → Überlagerung von linear polarisiertem und unpolarisiertem Licht
  - B) Minimale Intensität bei einer anderen Analysatorstellung als ohne $\lambda$/4-Plättchen
    → Überlagerung von elliptisch polarisiertem und unpolarisiertem Licht
